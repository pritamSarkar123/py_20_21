```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph

```python
x=[0,1,2,3,4,5]
y=[0,2,4,6,8,10]
#### resizing######
plt.figure(figsize=(8,5),dpi=100)## dpi-> pixels per inch

'''
plt.plot(x,y,label='2x',linewidth=2,
         marker='.',markersize=10,markeredgecolor='b',
         color='#ababab',## hex color->'#ababab'
         linestyle='-.', ##'--'
        ) 
'''
#### sorthand notation########'[color][marker][line style]' #CMS
plt.plot(x,y,'b^-.',label='2x')

x2=np.arange(0,4.5,0.5)
print(x2)
plt.plot(x2[:6],x2[:6]**2,'r',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'r',linestyle='--')

plt.title("Our First Graph",fontdict={'fontname':'Comic Sans MS',
                                      'fontsize':20})
plt.xlabel('XXXXX',fontdict={'fontname':'Arial',
                        'fontsize':15})
plt.ylabel('YYY',fontdict={'fontname':'Arial',
                        'fontsize':15})
plt.xticks([1,2,3,3.5,4],rotation='vertical')
plt.yticks(x2**2,rotation='vertical')
plt.legend()
plt.grid()
plt.savefig('mygraph.png')
plt.show()
```

# Barchart

```python
labels=['A','B','C']
values=[1,4,2]
plt.figure(figsize=(6,4),dpi=100)
bars=plt.bar(labels,values)
bars[0].set_hatch('/')
bars[1].set_hatch('O') #'o'
bars[2].set_hatch('*')
plt.show()
```

```python
labels=['A','B','C']
values=[1,4,2]
plt.figure(figsize=(6,4),dpi=100)
bars=plt.bar(labels,values)
patterns=['/','o','*']
for i,pattern in enumerate(patterns):
    bars[i].set_hatch(pattern)
plt.show()
```

```python
labels=['A','B','C']
values=[1,4,2]
plt.figure(figsize=(6,4),dpi=100)
bars=plt.bar(labels,values)
patterns=['/','o','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))## '0'<-pop from 1st item
                                ## other wise pop from back
plt.show()
```
