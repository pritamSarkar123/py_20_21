# Making changes

```python
df['Total']=df['HP']+df['Attack']+df['Defense']+df['Sp. Atk']+df['Sp. Def']+df['Speed']
#df.head()
df.drop(columns=['Total'],inplace=True)
df.head()
```

```python
df['Total']=df.iloc[:,4:10].sum(axis=1) #1 means horizontal addition
df.head()
```

```python
### Total before HP column
cols=list(df.columns.values) #list + num not alowed
df=df[cols[0:4]+[cols[-1]]+cols[4:12]] ## list + list + list ## []<-makes list
df.head()
```

```python
df.to_csv('modified.csv',index=False)
df.to_excel('modified.xlsx',index=False)
df.to_csv('modified.txt',index=False,sep='\t')
```

# Filtering Data

```python
## like select opetation in dbms
df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison')]
```

```python
df.loc[(df['Type 1']=='Grass') | (df['Type 2']=='Poison')]
```

```python
new_df=df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison') & (df['HP']>70)]
new_df.to_csv('filtered.csv',index=False)
```

```python
new_df=new_df.reset_index(drop=True) #setting new index and dropping old index
## new_df.reset_index(drop=True,inplace=True) does the same job
new_df
```

```python
#df.loc[~df['Name'].str.contains('Mega')] << opposite conditon ~ similar as !
df.loc[df['Name'].str.contains('Mega')]
```

```python
import re
###### if Type 1 is Fire or Grass
df.loc[df['Type 1'].str.contains('Fire|Grass',regex=True)]
```

```python
##flags=re.I    ignore case
df.loc[df['Type 1'].str.contains('fire|grass',flags=re.I,regex=True)]
```

```python
### if names starts with Pi   ---  ^pi[a-z]*
df.loc[df['Name'].str.contains('^pi[a-z]*',flags=re.I,regex=True)]
```

```python
### if names contains Pi   ---  pi[a-z]*
df.loc[df['Name'].str.contains('pi[a-z]*',flags=re.I,regex=True)]
```

# Conditional Changes

```python
### changing Fire to Flamer in Type 1 column

## checking condition on one column , make change on that column
df.loc[df['Type 1']=='Fire','Type 1']='Flamer'
df
```

```python
## all flamy pokemons are legendery 
## checking condition on one column , make change on another column
df.loc[df['Type 1']=='Flamer','Legendary']=True
df
```

```python
## based on a condition modifying multiple columns with same value
df=pd.read_csv('modified.csv')
df.loc[df['Total']>500,['Generation','Legendary']]='TEST VALUE'
df
```

```python
## based on a condition modifying multiple columns with diff value
df=pd.read_csv('modified.csv')
df.loc[df['Total']>500,['Generation','Legendary']]=['Test1','Test2']
df
```
