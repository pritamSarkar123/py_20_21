```python
import numpy as np
from itertools import permutations as perm
graph=[
    [1,1,0,1],
    [1,1,1,0],
    [0,1,1,1],
    [1,0,1,1]
]
#graph=[[1,1,1],[1,1,1],[1,1,1]]
graph=np.matrix(graph)
n=graph.shape[0]
colors=['B','R','G','Y','S']
colored_list=[]
for i in range(n):
    colored_list.append('_')
permuted_color=list(perm(colors,3))
permuted_color
```

```python
def possible(col,vertex):
    global n
    global graph
    global colored_list
    global color
    for v_check in range(n):
        ## adjacent vertes and same color ##
        if graph[vertex,v_check]==1 and colored_list[v_check]==col:
            return False
    return True
```

```python
def solve(v,color):
    global colored_list
    for c in color:
        if possible(c,v):
            colored_list[v]=c
            if v+1<n:
                solve(v+1,color)
            else:
                print(colored_list)
    
```

```python
def all_possible():
    global colored_list
    for color in permuted_color: 
        solve(0,color)
        ##### updating color list    
        colored_list=[]
        for i in range(n):
            colored_list.append('_')
```
